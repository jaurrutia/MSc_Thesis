### Example E Field

```
nDrude = Sqrt@DrudeEpsEnergy[hcPlanckSpeed/#,10., 0.015]&;

radius = 30.;
mesh = N[Array[{#1,0.,#2}&,{50,50}, 3. * radius *{{1,-1},{1,-1}}]];
dataxz = {{0,0},{0,0}};

lda = 342.;
indices = {1.33, nDrude[lda]};
AbsoluteTiming[ dataxz[[1,1]] = MieScatteredElectricField[mesh,indices, lda, radius, "Parallelize" -> True];]
AbsoluteTiming[ dataxz[[1,2]] = MieInternalElectricField[mesh,indices, lda, radius, "Parallelize" -> True]; ]


lda = 262.;
indices = {1.33, nDrude[lda]};
AbsoluteTiming[ dataxz[[2,1]] = MieScatteredElectricField[mesh,indices, lda, radius, "Parallelize" -> True];]
AbsoluteTiming[ dataxz[[2,2]] = MieInternalElectricField[mesh,indices, lda, radius, "Parallelize" -> True]; ]

Map[ListDensityPlot[Map[Chop@Norm, #,{-2}]]&, dataxz, {2}]
Map[ListDensityPlot[Map[Chop@Norm, #,{-2}]]&, Plus @@@ dataxz]
```
