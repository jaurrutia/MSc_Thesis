```
fs = 9;
Needs["MaTeX`"]

texStyle := {FontFamily -> "Latin Modern Roman", FontSize -> fs, Black};
graphsOpts := {Mesh -> Full, BaseStyle -> texStyle, Frame -> True, 
  FrameStyle -> Black, ImageSize -> 215, PlotStyle -> ColorData[3]}
SetOptions[ListLinePlot, graphsOpts];

nano = 1*^9;
```

```
nano = 1;
FileNames["*2.txt"]
data = Import[#,"Data"] &@@ %;
data = data /. {x_,y_,z_,E_} -> {y * nano,z * nano,E};
```

```
lim = 30
ListDensityPlot[data, PlotRange -> {{-lim,lim},{-lim,lim},{Automatic, Automatic}}, ColorFunction -> "Rainbow",
                PlotLegends -> Automatic,
                BaseStyle -> texStyle,
                 Frame -> True, 
                 FrameStyle -> Black, ImageSize -> 215,
                 FrameLabel -> (MaTeX[# <> "\\text{ [nm]}"]&/@{"x","z"})
                ]
Export["Isolated-COMSOL.pdf",%]
```
